```python
import sqlite3
import pickle
import os
import zipfile

database_path = './mydatabase.sqlite'

# connect to the database
conn = sqlite3.connect(database_path)
cursor = conn.cursor()

table_name = ["bond10", "'bond3-5'", "credit", "hs300", "maotai", "zz500"]
# execute the query statement
for item in table_name:
    print(item)
    cursor.execute("SELECT * FROM " + item)
    rows = cursor.fetchall()

    # transform results into pickle strings
    pickle_string = pickle.dumps(rows)

    filename = item + ".pickle"
    # write pickle strings into the file
    with open(filename, "wb") as f:
        f.write(pickle_string)

primary_file_path = "D:/pythonProject2/"
folder_path = "D:/pythonProject2/OS"
# zipfolder_path = "D:/pythonProject2/zipOS.zip"
zipfile_name = "zipOS.zip"

if not os.path.exists(folder_path):
    os.makedirs(folder_path)

# move pickle files into the specific folder
for item in table_name:
    file_path = primary_file_path + item + ".pickle"
    print(file_path)
    new_file_path = os.path.join(folder_path, os.path.basename(file_path))
    os.rename(file_path, new_file_path)

# open a new zip file in write mode
with zipfile.ZipFile(zipfile_name, 'w', zipfile.ZIP_DEFLATED) as zipf:
    # iterate through all the files and folders in the folder
    for root, _, files in os.walk(folder_path):
        for file in files:
            # construct the full file path and the corresponding archive name
            file_path = os.path.join(root, file)
            archive_name = os.path.relpath(file_path, folder_path)
            # add the file to the zip file with the archive name
            zipf.write(file_path, archive_name)

# delete all files generated before
for item in table_name:
    os.remove(folder_path + "/" + item + ".pickle")
os.rmdir(folder_path)

# close the connection with database
cursor.close()
conn.close()

# specify the path of the compressed file
compressed_file_path = primary_file_path + zipfile_name

# specify the directory where you want to extract the files
extracted_folder_path = "extracted_folder"

# open the compressed file
with zipfile.ZipFile(compressed_file_path, 'r') as zip_ref:
    # extract all files to the specified directory
    zip_ref.extractall(extracted_folder_path)

# the files have been extracted to the specified directory
```
